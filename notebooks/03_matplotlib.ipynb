**<u>THEORY PART AND INFO ABOUT MATPLOTLIB<u>**

Matplotlib is a library used for creating static, animated, and interactive visualizations. It is especially useful for generating plots, charts, and other graphical representations of data.

```python
import matplotlib.pyplot as plt # importing the library for plotting
```

```python
import numpy as np # importing the library for numerical operations

x = np.array(np.arange(100)) # creating an array of 100 numbers from 0 to 99
y = 2*x 

plt.plot(x, y) # plotting the points (x, y) on a graph
plt.title('Linear model') # adding a title to the graph
```

```python
y = 3*x + 2
plt.plot(x, y, c='r') # Red line
y = -2*x + 6
plt.plot(x, y, c='c', linewidth=3) # Cyan line with thicker width
y = -9*x
plt.plot(x, y, c='m', marker='+', markersize=5) # Magenta line with plus markers

plt.legend(['Linear model 1', 'Linear model 2', 'Linear model 3'])
```

You can use subplots to display multiple plots side by side or in a grid. This is especially useful in Machine Learning when visualizing different variables, model outputs, or evaluation results.

```python
x = np.linspace(0, 10, 100) # imput data

y_true = 3 * x + 5 # true function

y_pred = 2.8 * x + 4.5 # predicted function

error = y_true - y_pred # calculating the error between true and predicted values

fig, axs = plt.subplots(1, 3, figsize=(15, 4)) # creating a figure with 1 row and 3 columns of subplots

axs[0].plot(x, y_true, label='True', color='blue') # plotting the true function on the first subplot
axs[0].plot(x, y_pred, label='Predicted', color='orange', linestyle='--') # plotting the predicted function on the first subplot with dashed line
axs[0].set_title('True vs Predicted') # adding a title to the first subplot
axs[0].legend() # adding a legend to the first subplot

axs[1].plot(x, y_pred, color='green') # plotting the predicted function on the second subplot with green color
axs[1].set_title('Model Prediction') # adding a title to the second subplot

axs[2].plot(x, error, color='red') # plotting the error on the third subplot with red color
axs[2].set_title('Prediction Error') # adding a title to the third subplot

plt.tight_layout() # adjusting the layout to prevent overlap between subplots
plt.show() # displaying the figure with all subplots
```
